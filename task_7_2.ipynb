```python
def check_input(element):
    return isinstance(element, str)


def get_substring(s):
    max_sub = ""
    current = ""
    used = set()

    left = 0
    for right in range(len(s)):
        while s[right] in used:
            used.remove(s[left])
            left += 1
        used.add(s[right])
        current = s[left:right + 1]
        if len(current) > len(max_sub):
            max_sub = current

    return max_sub


def main(lst):
    max_len = 0
    max_index = -1
    max_string = ""

    for i, elem in enumerate(lst):
        if check_input(elem):
            if len(elem) > max_len:
                max_len = len(elem)
                max_index = i
                max_string = elem

    if max_index == -1:
        print(-1, "")
        return

    result_substring = get_substring(max_string)
    print(max_index, result_substring)
```
